```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB,ComplementNB
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import pickle
```

```python
from processing.pre_processing import x_train,x_test,y_train,y_test
```

```python
lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
pred=lr_model.predict(x_test)
pred
```

```python
lr_model.score
```

```python
print(accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
cm=confusion_matrix(y_test,pred)
cm
```

```python
rfc=RandomForestClassifier(n_estimators=15, random_state=10)
rfc.fit(x_train, y_train)
```

```python
rfc_pred=rfc.predict(x_test)
rfc_pred
```

```python
cm=confusion_matrix(y_test, rfc_pred)
cm
```

```python
rfc_score=accuracy_score(y_test,rfc_pred)
rfc_score
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5)
gb_clf.fit(x_train, y_train)
```

```python
pred2 = gb_clf.predict(x_test)
pred2
```

```python
accuracy = accuracy_score(y_test, pred2)
accuracy
```

```python
svm_clf = SVC(kernel='rbf', C=1, gamma='scale')
svm_clf.fit(x_train, y_train)
```

```python
y_pred = svm_clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
